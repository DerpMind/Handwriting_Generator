```python
pxls = []
for bad_example in fonts:
    downsize_factor = downsize_parameters(bad_example, 'J')
    x_shift, y_shift = decrop_parameters(bad_example,'J', downsize_factor=downsize_factor)
    dilate_factor = dilate_parameters(bad_example,'J', downsize_factor=downsize_factor,
                                     x_shift=x_shift, y_shift=y_shift)
    x_shift, y_shift = center_parameters(bad_example, 'J', downsize_factor=downsize_factor,
                                        dilate_factor=dilate_factor, x_shift=x_shift, y_shift=y_shift)
    pxl = processed_img_constructor(bad_example, 'J', downsize_factor=downsize_factor,
                                   dilate_factor=dilate_factor, x_shift=x_shift, y_shift=y_shift)
    pxls.append(pxl)
    
```

```python
for i in pxls:
    plt.figure(figsize=(5,5))
    plt.imshow(i)
```
