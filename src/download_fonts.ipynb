```python
#https://www.dafont.com/theme.php?cat=603&page=48&fpp=100
import requests
from bs4 import BeautifulSoup
import pickle
```

```python
def scrape_one_page(html_page_souped):
    dlboxes = html_page_souped.findAll("div", {"class": "dlbox"})
    dllinks = [div.find('a')['href'] for div in dlboxes]
    return dllinks
```

```python
import random, time
from itertools import chain

def get_download_links(max_page=5):
   
    list_of_links = []
    for i in range(max_page):
        page_no = i+1
        link = f"https://www.dafont.com/theme.php?cat=603&page={page_no}&fpp=100"
        page = requests.get(link, headers={'User-Agent':'test'})
        page_souped = BeautifulSoup(page.text, 'html.parser')
    
        next_links = scrape_one_page(page_souped)
        list_of_links.append(next_links)
        
        time.sleep(random.randint(1, 10)/10)
    
    list_of_links = list(chain.from_iterable(list_of_links))
    return list_of_links
```
